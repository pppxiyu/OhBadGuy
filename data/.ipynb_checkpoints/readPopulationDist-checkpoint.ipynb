```python
from raster2xyz.raster2xyz import Raster2xyz


input_raster = "populationDist.tif"
out_csv = "populationDist.csv"

rtxyz = Raster2xyz()
rtxyz.translate(input_raster, out_csv)
```

```python
import pandas as pd 
myRasterDF = pd.read_csv('populationDist.csv')
myRasterDF
```

```python
import geopandas as gpd
```

```python
gdf = gpd.GeoDataFrame(
    myRasterDF[:100], geometry = gpd.points_from_xy(myRasterDF[:100].x, myRasterDF[:100].y))
```

```python
gdf = gdf.set_crs('ESRI:54009').to_crs('EPSG:4326')
gdf
```
